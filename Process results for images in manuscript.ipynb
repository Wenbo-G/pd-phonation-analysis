# Results summary

```python
import dataLoader as dl
import resultsHandler as rh
from os.path import join
import plotly.graph_objects as go
import pandas as pd
import numpy as np

x = ['Unmodified','Digital fingerprinting removed','Digital fingerprinting and accuracy paradox removed','All removed']

accuracies = pd.DataFrame(columns=['OPD','mPower'],dtype=object)
stdevs = pd.DataFrame(columns=['OPD','mPower'],dtype=object)


best_ozkan_model_OPD = 'ozkan PCA_14 k_1 MinMaxScaler X only'
best_caliskan_model_OPD = 'caliskan ReLU Sigmoid latent:6, epochs:400, lr:0.0030 MinMaxScaler X only'
best_ulhaq_model_OPD = 'ulhaq rbf gamma:0.2000 C:5 num_features:14 StandardScaler X only'

best_ozkan_model_mPower = 'ozkan PCA_16 k_11 StandardScaler X only'
best_caliskan_model_mPower = 'caliskan ReLU Sigmoid latent:6, epochs:50, lr:0.0030 StandardScaler X only'
best_ulhaq_model_mPower = 'ulhaq rbf gamma:scale C:10 num_features:20 StandardScaler X only'
```

```python
for best_model,name in zip([best_ozkan_model_OPD,best_caliskan_model_OPD,best_ulhaq_model_OPD],['A','B','C']):
    accuracies.at['kf %s' % name,'OPD'] = []
    stdevs.at['kf %s' % name,'OPD'] = []
    accuracies.at['split %s' % name,'OPD'] = []
    stdevs.at['split %s' % name,'OPD'] = []
        
for best_model,name in zip([best_ozkan_model_mPower,best_caliskan_model_mPower,best_ulhaq_model_mPower],['A','B','C']):
    accuracies.at['kf %s' % name,'mPower'] = []
    stdevs.at['kf %s' % name,'mPower'] = []
    accuracies.at['split %s' % name,'mPower'] = []
    stdevs.at['split %s' % name,'mPower'] = []
```

```python
#Load the unmodified (replicated) results
kf_OPD_0 = dl.pickleLoad(join('Results','repeated_kfold with OPD unmodified 27November21.pickle'))
split_OPD_0 = dl.pickleLoad(join('Results','repeated_traintest with OPD unmodified27November21.pickle'))
kf_mPower_0 = dl.pickleLoad(join('Results','repeated_kfold with mPower unmodified 24January22.pickle'))
split_mPower_0 = dl.pickleLoad(join('Results','repeated_traintest with mPower unmodified24January22.pickle'))


#Get best results models for each
for best_model,name in zip([best_ozkan_model_OPD,best_caliskan_model_OPD,best_ulhaq_model_OPD],['A','B','C']):
    accuracies.at['kf %s' % name,'OPD'].append(kf_OPD_0.loc[best_model,'average accuracy']*100)
    stdevs.at['kf %s' % name,'OPD'].append(kf_OPD_0.loc[best_model,'std accuracy']*100)
    accuracies.at['split %s' % name,'OPD'].append(split_OPD_0.loc[best_model,'average accuracy']*100)
    stdevs.at['split %s' % name,'OPD'].append(split_OPD_0.loc[best_model,'std accuracy']*100)
    
    
for best_model,name in zip([best_ozkan_model_mPower,best_caliskan_model_mPower,best_ulhaq_model_mPower],['A','B','C']):
    accuracies.at['kf %s' % name,'mPower'].append(kf_mPower_0.loc[best_model,'average accuracy']*100)
    stdevs.at['kf %s' % name,'mPower'].append(kf_mPower_0.loc[best_model,'std accuracy']*100)
    accuracies.at['split %s' % name,'mPower'].append(split_mPower_0.loc[best_model,'average accuracy']*100)
    stdevs.at['split %s' % name,'mPower'].append(split_mPower_0.loc[best_model,'std accuracy']*100)
```
